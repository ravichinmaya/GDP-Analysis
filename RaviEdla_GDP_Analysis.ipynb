```python
df_P1A.columns = df_P1A.columns.str.replace(' ', '_')
df_P1A.columns = df_P1A.columns.str.rstrip('_')
df_P1A
```

```python
## Selcting rows of % Growth over previous year
df_GDP1A=df_P1A.loc[df_P1A["Items__Description"] == "(% Growth over previous year)"]
df_GDP1A['Duration'] = df_GDP1A['Duration'].str[0:4].astype(int)
df_GDP1A.head(20)
```

```python
##best fit line to represent the growth for each state
gdp_each_state= df_GDP1A.drop('All_India_GDP', axis=1)
gdp_each_state = gdp_each_state.T

gdp_each_state = gdp_each_state[3:]
gdp_each_state = gdp_each_state.apply(lambda x:x.fillna(x.mean()))

#print(gdp_each_state)
#gdp_each_state = gdp_each_state.reset_index()

for i,j in gdp_each_state.iterrows():
    state_name = i
    state_value = j.astype(int)
    num = np.array([1,2,3,4,5], dtype=np.int)
    
    plt.figure() 
    
    plt.xlabel("Duration");
    plt.ylabel(state_name)
    m, c=np.polyfit(num,state_value,1)
    plt.scatter(x ='Duration',y=state_name, data = i)
    plt.plot(num, m*num+c, '--b')
```

```python
#best fit line graph representing the Growth of the Nation over period of time 2012-2013 to 2016-2017

plt.xlabel("Duration");plt.ylabel("All_India_GDP")
m, c=np.polyfit(df_GDP1A['Duration'],df_GDP1A['All_India_GDP'],1)
plt.scatter(x ='Duration',y='All_India_GDP', data = df_GDP1A)
plt.plot(df_GDP1A['Duration'], m*df_GDP1A['Duration']+c, '--g')
```

```python
#growth rate of your home state

y=df_GDP1A["Telangana"];
x=df_GDP1A["Duration"];
plt.xlabel("Duration");plt.ylabel("Telangana")
plt.scatter(x,y)
```
